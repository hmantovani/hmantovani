### Importando os arquivos e defindo os índices.

```python
jogos2013 = pd.read_csv('jogos_2013.csv', index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
jogos2014 = pd.read_csv('jogos_2014.csv', index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
jogos2015 = pd.read_csv('jogos_2015.csv', index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
jogos2016 = pd.read_csv('jogos_2016.csv', index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
jogos2017 = pd.read_csv('jogos_2017.csv', index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
estados = pd.read_csv('estados.csv', index_col=['id_estado'])
arenas = pd.read_csv('arenas.csv', index_col=['id_arena'])
clubes = pd.read_csv('clubes.csv', index_col=['id_clube'])
```

```python
jogos2013.head()
```

### Mesclando os arquivos de jogos em um só.

```python
# CONCATENANDO OS ANOS
jogos = pd.concat([jogos2013, jogos2014, jogos2015, jogos2016, jogos2017], axis=0)
jogos.head()
```

### Desnomalizando os arquivos.

```python
# JUNÇÃO ENTRE OS JOGOS E OS NOMES DAS ARENAS
jogos = jogos.join(arenas, how='left', on=['id_arena'])
jogos.head()
```

```python
# JUNÇÃO ENTRE OS JOGOS E OS NOMES DE CLUBES 1 E 2
# O MÉTODO JOIN USA COMO DEFAULT HOW=LEFT, PORTANTO NÃO PRECISAMOS DEFINIR DAQUI EM DIANTE...
jogos = jogos.join(clubes, on=['id_clube1'], lsuffix='1', rsuffix='2')
jogos = jogos.join(clubes, on=['id_clube2'], lsuffix='1', rsuffix='2')
jogos.head()
```

```python
# JUNÇÃO ENTRE OS JOGOS E OS NOMES DOS CLUBES VENCEDORES
jogos = jogos.join(clubes, on=['id_clube_vencedor'])
jogos.head()
```

```python
# POR PADRÃO O NOME DO CAMPO SERÁ INCLUÍDO DO DF SERÁ EXATAMENTE COMO ESTÁ NO DF ORIGINAL
# VAMOS RENOMEAR PARA FICAR CLARO O QUE QUEREMOS DIZER...
jogos.rename(columns={'desc_clube':'desc_clube_vencedor'}, inplace=True)
jogos.head()
```

```python
# JUNÇÃO ENTRE OS JOGOS E OS NOMES DE DOS ESTADOS DOS CLUBES 1 E 2
jogos = jogos.join(estados, on=['id_estado_clube1'], lsuffix='1', rsuffix='2')
jogos = jogos.join(estados, on=['id_estado_clube2'], lsuffix='1', rsuffix='2')
jogos.head()
```

```python
# JUNÇÃO ENTRE OS JOGOS E OS NOMES DOS ESTADOS VENCEDORES
jogos = jogos.join(estados, on=['id_estado_vencedor'])
jogos.rename(columns={'desc_estado':'desc_estado_vencedor', 'sigla':'sigla_vencedor'}, inplace=True)
jogos.head()
```

```python
# A PARTI DE AGORA NÃO PRECISAMOS MAIS DAS CONSULAS DE INDICE, VAMOS ELIMINÁ-LAS A PARTI DAQUI.
jogos.reset_index(drop=True, inplace=True)
```

```python
jogos.head()
```

### Explorando os dados.

#### a. Qual o clube que mais marcou gols em todas as temporadas?

```python
# PRIMEIRO PRECISAMOS SABER SE O CLUBE QUE MARCOU MAIS GOLS ESTÁ COMO clube1 OU clube2
print("clube 1, maior qtd de gols:", jogos.groupby('desc_clube1')['qtd_gols_clube1'].sum().max())
print("clube 2, maior qtd de gols:", jogos.groupby('desc_clube2')['qtd_gols_clube2'].sum().max())
```

```python
# AGORA QUE SABEMOS QUE FOI O CLUBE LOCALIZADO NO CAMPO clube1, PODEMOS MOSTRA-LO
jogos.groupby('desc_clube1')['qtd_gols_clube1'].sum().nlargest(1)
```

```python
# PODEMOS FAZER ISSO DE FORMA MAIS DINÂMICA, CONFORME ABAIXO:
if jogos.groupby('desc_clube1')['qtd_gols_clube1'].sum().max() > jogos.groupby('desc_clube2')['qtd_gols_clube2'].sum().max():
    print(jogos.groupby('desc_clube1')['qtd_gols_clube1'].sum().nlargest(1))
else:
    print(jogos.groupby('desc_clube2')['qtd_gols_clube2'].sum().nlargest(1))
```

#### b. Qual o clube que mais marcou gols na temporada de 2015?

```python
# NESSE CASO, PRIMEIRO PRECISAMOS RESTRINGIR OS RESULTADOS APENAS PARA REGISTROS DE 2015
# COMO TEMOS UMA DATA COMPLETA, PODEMOS USAR O MÉTODO slice() PARA BUSCAR APENAS A FRAÇÃO DE DADOS REFERENTE AO ANO
jogos[jogos['data'].str.slice(start=6) == '2015']
```

```python
# FEITO ISSO, PODEMOS APLICAR UMA LÓGICA SEMELHANTE AO QUE FIZEMOS NO EXERCÍCIO ANTERIOR
print("clube 1, maior qtd de gols em 2015:", jogos[jogos['data'].str.slice(start=6) == '2015'].groupby('desc_clube1')['qtd_gols_clube1'].sum().max())
print("clube 1, maior qtd de gols em 2015:", jogos[jogos['data'].str.slice(start=6) == '2015'].groupby('desc_clube2')['qtd_gols_clube2'].sum().max())
```

```python
# RESULTADO
jogos[jogos['data'].str.slice(start=6) == '2015'].groupby('desc_clube1')['qtd_gols_clube1'].sum().nlargest(1)
```

```python
# TORNANDO MAIS DINÂMICO
if jogos[jogos['data'].str.slice(start=6) == '2015'].groupby('desc_clube1')['qtd_gols_clube1'].sum().max() > jogos[jogos['data'].str.slice(start=6) == '2015'].groupby('desc_clube2')['qtd_gols_clube2'].sum().max():
    print(jogos[jogos['data'].str.slice(start=6) == '2015'].groupby('desc_clube1')['qtd_gols_clube1'].sum().nlargest(1))
else:
    print(jogos[jogos['data'].str.slice(start=6) == '2015'].groupby('desc_clube2')['qtd_gols_clube2'].sum().nlargest(1))
```

#### c. Qual o jogo onde houveram mais gols, em todas as temporadas?

```python
# PRIMEIRO VAMOS CRIAR UM NOVO CAMPO QUE MOSTRA O TOTAL DE GOLS
jogos['qtd_total_partida'] = jogos['qtd_gols_clube1'] + jogos['qtd_gols_clube2']
# FEITO ISSO, PODEMOS ORDENAR OS VALORES DE FORMA DECRESCENTE PARA SABER QUAL FOI O JOGO
jogos.sort_values(by=['qtd_total_partida'], ascending=False).head(1)
```

#### d. Qual foi a maior goleada de todas as temporadas?

```python
# GOLEADA, QUER DIZER A MAIOR DIFERENÇA ENTRE OS GOLS DE UM TIME EM RELAÇÃO AO OUTRO
# VAMOS SUBTRAIR A QUANTIDADE DE GOLS DE UM CLUBE PELO OUTRO
jogos['qtd_abs_goleada'] = jogos['qtd_gols_clube1'] - jogos['qtd_gols_clube2']
# FEITO ISSO PODEMOS APLICAR O MÉTODO PARA OBTER VALORES ABSOLUTOS
jogos['qtd_abs_goleada'] = jogos['qtd_abs_goleada'].abs()
jogos.sort_values(by=['qtd_abs_goleada'], ascending=False).head(1)
```

#### e. Qual foi o estado mais vitorioso (quantidade de partidas) em 2016?

```python
jogos[jogos['data'].str.slice(start=6) == '2016'].groupby('desc_estado_vencedor')['data'].count().sort_values(ascending=False)
```
